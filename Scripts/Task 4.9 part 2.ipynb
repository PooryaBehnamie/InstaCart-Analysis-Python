### 2. Sampling Data

```python
# creating a list holding True/False values to the test np.random.rand() <= 0.7
np.random.seed(4)
dev = np.random.rand(len(df)) <= 0.7
dev
```

```python
# store 70 % of the sample in big data frame
big = df[dev]
# store 30 % of the sample in small data frame
small = df[~dev]
```

### 3. Creating Bar Charts, Histogram, scatterplot, Line chart

```python
# Creat a bar chart:
bar = df['orders_day_of_week'].value_counts().sort_index().plot.bar(color=['navy','blue', 'purple', 'orange', 'yellow', 'red', 'gray'])
```

```python
# Creat a Histogram
hist = df['prices'].plot.hist(bins = 20)
```

```python
# creat a scatter plot
scat = sns.scatterplot(x = 'prices', y = 'prices', data = df)
```

```python
# Creat a Line Chart using Sample
df_2 = small[['orders_day_of_week','prices']]
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
# Exporting Figures
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
hist.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hist_prices.png'))
scat.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'scat_prices.png'))
line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_order_dow_prices.png'))
```
