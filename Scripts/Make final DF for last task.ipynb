### Memory optimization

```python
# Change some columne data type to optimize memory
df['aisle_id'] = df['aisle_id'].astype('int32')
df['department_id'] = df['department_id'].astype('int32')
df_dep['department_id'] = df_dep['department_id'].astype('int32')
df['prices'] = df['prices'].astype('float32')
df['avg_spent'] = df['avg_spent'].astype('float32')
df['age'] = df['age'].astype('int32')
df['n_dependants'] = df['n_dependants'].astype('int32')
```

### Combine two Dataframes

```python
# Combine two data frames 
df = df.merge(df_dep, on = 'department_id', indicator = True)
```

```python
df['_merge'].value_counts(dropna=False)
```

```python
df.drop(columns = '_merge', inplace = True)
```

```python
# Make a new dataframe to change days of the week column values to their name
dict = {'orders_day_of_week':[0, 1, 2, 3, 4, 5, 6], 'orders_day':['Saturday', 'Sunday', 'Monday','Tuesday', 'Wednesday', 'Thursday', 'Friday']}
df_days = pd.DataFrame(dict)
df_days
```

```python
# Combine two data frames
df = df.merge(df_days, on = 'orders_day_of_week', indicator = True)
```
